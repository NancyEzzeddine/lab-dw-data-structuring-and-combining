## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here
import pandas as pd
url_file1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url_file2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url_file3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

file1 = pd.read_csv(url_file1)
file2 = pd.read_csv(url_file2)
file3 = pd.read_csv(url_file3)

combined_data = pd.concat([file1, file2, file3], ignore_index=True)

combined_data.head()
```

```python

# drop duplicated columngs
combined_data_cleaned = combined_data.drop(columns=['State', 'Gender'])

# clean 'Customer Lifetime Value' by removing '%' and converting to numeric
combined_data_cleaned['Customer Lifetime Value'] = combined_data_cleaned['Customer Lifetime Value'].astype(str).str.replace('%', '')
combined_data_cleaned['Customer Lifetime Value'] = pd.to_numeric(combined_data_cleaned['Customer Lifetime Value'], errors='coerce')


# fix number of opern complaints
def clean_open_complaints(value):
    if isinstance(value, str):
        return int(value.split('/')[1])
    elif pd.isna(value):
        return 0  
    else:
        return int(value)

combined_data_cleaned['Number of Open Complaints'] = combined_data_cleaned['Number of Open Complaints'].apply(clean_open_complaints)


# handle missing values
combined_data_cleaned['GENDER'].fillna(combined_data_cleaned['GENDER'].mode()[0], inplace=True)

combined_data_cleaned['Customer Lifetime Value'].fillna(combined_data_cleaned['Customer Lifetime Value'].median(), inplace=True)

combined_data_cleaned
```

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here

file_path = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv' 
data_new = pd.read_csv(file_path)

print(data_new.head())

print(data_new.info())

print(data_new.describe())

print(data_new.isnull().sum())
```

```python
data_new.drop(columns=['unnamed:_0'], inplace=True)


data_new.columns = data_new.columns.str.lower().str.replace(' ', '_')


data_new['effective_to_date'] = pd.to_datetime(data_new['effective_to_date'])


data_new['month'] = data_new['month'].astype('category')


print(data_new.describe())
```

```python

clv_gender_education_summary = data_new.pivot_table(
    values='customer_lifetime_value',
    index='gender',
    columns='education',
    aggfunc='mean'
).round(2)


clv_gender_education_summary
```
